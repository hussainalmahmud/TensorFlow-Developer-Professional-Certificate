```python
import tensorflow as tf
from tensorflow import keras


from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

sentences = [
    'I love my dog',
    'I love my cat',
    'You love my dog!',
    'Do you think my dog is amazing?'
]

tokenizer = Tokenizer(num_words = 100, oov_token="<OOV>") # 100 is to take the top 100 common words, oov_token to fill the gap with a special word
tokenizer.fit_on_texts(sentences) #fit the data to tokenzier
word_inde = tokenizer.word_index #call the tokenizer object to get the word indexes

sequences = tokenizer.texts_to_sequences(sentences) # change the text to sequences

padded = pad_sequences(sequences, maxlen=5) # add padding to make the sentences length the same
print("\nWord Index = " , word_inde) 
print("\nSequences = " , sequences)
print("\nPadded Sequences:")
print(padded)


# Try with words that the tokenizer wasn't fit to
test_data = [
    'i really love my dog',
    'my dog loves my manatee'
]

test_seq = tokenizer.texts_to_sequences(test_data) # fit the test data
print("\nTest Sequence = ", test_seq)
padded = pad_sequences(test_seq, maxlen=10)
print("\nPadded Test Sequence: ")
print(padded)
```
